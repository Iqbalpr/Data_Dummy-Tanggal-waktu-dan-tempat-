```python
import pandas as pd
import io 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import datetime
```

```python
url = 'https://raw.githubusercontent.com/Iqbalpr/Data_Dummy-Tanggal-waktu-dan-tempat-/main/Dummy%20Per%20Jam%20-%20Sheet1.csv'
Data = pd.read_csv(url)
Data
```

# **Point Of Interest**

```python
#Melihat Destinasi Awal
x = Data['Date']
y = Data['Place']
plt.figure(figsize=(16,12))
plt.title('Tempat asal (Rumah)')
plt.scatter(x,y)
plt.show
```

Dapat dilihat dari grafik scatter plot diatas, bahwa pada tanggal 1-7 Maret 2022, tempat awal (Rumah) atau memulai hari adalah didaerah **Utan Kayu Selatan** lalu melanjutkan ke daerah **pasar minggu, tanjung barat dan balik lagi ke arah pasar minggu dan kembali ke Rumah**, sesuai dengan tabel "Data" yang disajikan sebelumnya

```python
# Untuk melihat waktu yang dihabiskan disuatu tempat
y = Data['Time']
x = Data['Place']
plt.figure(figsize=(16,12))
plt.title('Waktu yang dihabiskan disuatu tempat')
plt.scatter(x,y)
plt.show
# Untuk melihat berapa jam yang dihabiskan disuatu tempat selama 7 hari
Data['Place'].value_counts()
```

Dapat dilihat dari grafik scatter plot, bahwa waktu yang paling banyak dihabiskan adalah di Tanjung Barat dari **jam 8:00:00 - jam 16:00:00 (8 jam)** dan mempunyai waktu kunjungan sebesar **45 Kali**. Sehingga dapat disimpulkkan bahwa **Tanjung Barat** adalah tempat bekerja

```python
Data['Time'].value_counts()
```

dan untuk mencari tau berapa lama waktu yang telah dihabiskan selama 7 hari di tempat kerja, kita dapat melihat bahwa dari jam 08:00:00 - 16:00:00 terdapat 7 kunjungan, sehingga telah menghabiskan waktu paling banyak **56 jam (8*7)** ditempat kerja selama 7 hari 

# **Rute Favorit**

```python
Graf = nx.from_pandas_edgelist(Data, 'Date', 'Place', edge_attr=True, create_using=nx.MultiDiGraph())
```

```python
plt.figure(figsize=(12,12))

pos = nx.spring_layout(Graf)
nx.draw(Graf, with_labels=True, node_color="skyblue", edge_cmap=plt.cm.Blues, pos=pos)

plt.show()
```

Dengan bantuan Graf di atas, menunjukan rute terfavorit yang sering dilewati. Semakin tebal garis maka menandakan rute yang paling banyak dilewati adalah **Utan Kayu Selatan**, **Pasar Minggu** dan **Tanjung Barat**
